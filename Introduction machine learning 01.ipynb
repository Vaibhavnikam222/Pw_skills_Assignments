# Q.2  What is supervised learning? List some example supervised learning?

Supervised learning is a type of machine learning where the algorithm learns from a labeled dataset, which means the training data consists of input-output pairs. In other words, each example in the dataset is associated with a **known target or output**, and the algorithm's task is to learn a mapping from the input data to the correct output. 

The goal is to make predictions or classifications based on this learned mapping when presented with new, unseen data.

### Example:
1. sentiment analysis models can classify text data (such as social media posts or product reviews) into categories like positive, negative, or neutral.
2. Banks and financial institutions use supervised learning to predict the creditworthiness of applicants, deciding whether to approve or reject a loan application based on various factors.
3. Predicting the future stock prices of companies based on historical financial data and market indicators involves regression
4. Identifying objects or features within images

# Q. 3 What is unsupervised learning? List some example of unsupervised learning?

Unsupervised learning is a type of machine learning where the algorithm is trained on unlabeled data, meaning it doesn't have access to predefined output labels or target values. Instead, the algorithm's objective is to discover patterns, structures, or relationships within the data without explicit guidance.

Unsupervised learning is often used for tasks such as clustering and dimensionality reduction.

### Example:
1. K-Means Clustering: This algorithm is used to partition data into clusters based on similarity. It's widely applied in market segmentation, where customers are grouped into segments based on their purchasing behavior.
2. Anomaly Detection: Unsupervised learning can be used to detect anomalies or outliers in data, such as identifying fraudulent transactions in a credit card dataset where most transactions are legitimate.
3. Image Compression: Unsupervised learning can be applied to reduce the size of images while preserving important visual information, making them easier to store and transmit.

# Q. 5 What is main difference between supervised, unsupervised, and semi- supervised learning?

### Supervised Learning:

Labeled Data: In supervised learning, the algorithm is trained on a labeled dataset, where each example in the training data is associated with a known output or target.
Objective: The primary goal of supervised learning is to learn a mapping from input features to the correct output labels. It aims to make predictions or classifications based on this learned mapping.
Examples: spam email detection, image categorization , predicting house prices, estimating patient risk.

### Unsupervised Learning:

Unlabeled Data: Unsupervised learning algorithms are trained on unlabeled data, which means there are no predefined output labels or targets. The algorithm's goal is to discover patterns, structures, or relationships within the data.
Objective: Unsupervised learning is used for exploratory data analysis, clustering similar data points together, dimensionality reduction, density estimation, and other tasks that don't involve predicting specific output labels.
Examples:  customer segmentation, document clustering, dimensionality reduction , and anomaly detection are common unsupervised learning tasks.

### Semi-Supervised Learning:

Mixed Data: Semi-supervised learning combines elements of both supervised and unsupervised learning. It leverages a dataset that contains both labeled and unlabeled examples.
Objective: The main objective in semi-supervised learning is to improve the model's performance by utilizing the information contained in the limited labeled data while also leveraging the larger pool of unlabeled data for generalization.
Semi-supervised learning can be beneficial when labeled data is scarce or expensive to obtain.
Examples: Applications include text classification with a small labeled dataset and a large pool of unlabeled text documents, or image classification when only a subset of images has been manually labeled.

# Q.6  What is train,test, validation split? Explain importance of each term.

1. Training Set:

It is the largest portion of the dataset, and it is used to train the machine learning model. During training, the model learns the underlying patterns and relationships in the data by adjusting its internal parameters.

**Importance:** The training set is crucial because it is where the model learns from labeled examples. It helps the model generalize from the patterns it observes in the training data so that it can make predictions or classifications on new, unseen data. The quality and representativeness of the training data significantly impact the model's performance.

2. Test Set:

It is a separate portion of the dataset that is not used during training. It is reserved exclusively for evaluating the model's performance. After the model has been trained, it is tested on the test set to assess how well it generalizes to new, unseen data.

**Importance:** The test set provides an unbiased assessment of the model's performance. It helps determine whether the model has learned the underlying patterns in the data without overfitting  or underfitting. The test set's results give an estimate of how the model will perform in real-world scenarios.

3. Validation Set:

It is an optional subset of the dataset, and its primary purpose is to fine-tune the model's hyperparameters. Hyperparameters are configuration settings that are not learned during training but affect the model's behavior, such as learning rates, regularization strengths, and the number of hidden layers in a neural network.

**Importance:** The validation set allows the model developer to experiment with different hyperparameter settings and assess their impact on the model's performance. It helps in finding the best hyperparameter values that optimize the model's accuracy and generalization capabilities. Without a validation set, it's challenging to fine-tune a model effectively, and the risk of overfitting the hyperparameters to the test set is high.

# Q.7 How can unsupervised learning be used in anamoly detection?

Unsupervised machine learning can be used in anomaly detection by leveraging the fact that anomalies are rare and unusual events that deviate significantly from the normal behavior of a system. Here are some common techniques used for anomaly detection using unsupervised machine learning:

1. Clustering: Clustering algorithms such as k-means, DBSCAN, and hierarchical clustering can be used to group similar data points together. Anomalies can then be identified as data points that do not belong to any of the clusters or are in small, isolated clusters.

2. Density-Based Outlier Detection: Density-based techniques such as Local Outlier Factor (LOF) and Isolation Forest can be used to identify outliers that have a low density compared to their neighbors.

3. Dimensionality Reduction: Dimensionality reduction techniques such as Principal Component Analysis (PCA) and t-SNE can be used to reduce the high-dimensional data into a lower-dimensional space. Anomalies can then be identified as data points that are far from the bulk of the data in the reduced space.

4. Autoencoders: Autoencoders are a type of neural network that can be used for unsupervised anomaly detection. The autoencoder learns to reconstruct the input data, and anomalies can be identified as data points that are poorly reconstructed.

# Q.8 List down some commonly used supervised learning algorithm and unsupervised learning algorithm. 

### Supervised Learning Algorithms:

1. Linear Regression: Used for regression tasks, it models the relationship between input features and a continuous target variable.

2. Logistic Regression: Used for binary classification, it models the probability of a binary outcome.

3. Decision Trees: Builds a tree structure to make decisions and classify data points based on feature values.

4. Random Forest: An ensemble method that combines multiple decision trees to improve accuracy and reduce overfitting.

5. Support Vector Machines (SVM): Used for both classification and regression, SVM finds a hyperplane that best separates data points in a high-dimensional space.

6. K-Nearest Neighbors (KNN): A simple classification algorithm that assigns a class label based on the majority class among its k-nearest neighbors.

7. Naive Bayes: A probabilistic classifier that is based on Bayes' theorem and is often used for text classification and spam detection.

8. Neural Networks: Deep learning models composed of interconnected layers of artificial neurons. They are used for various tasks like image recognition and natural language processing.


### Unsupervised Learning Algorithms:

1. K-Means Clustering: Divides data points into clusters based on similarity, where each point belongs to the cluster with the nearest mean.

2. Hierarchical Clustering: Builds a hierarchy of clusters, typically represented as a dendrogram.

3. Principal Component Analysis (PCA): A dimensionality reduction technique that transforms data into a new coordinate system to reduce dimensionality while preserving as much variance as possible.

4. t-Distributed Stochastic Neighbor Embedding (t-SNE): A dimensionality reduction technique known for its effectiveness in visualization.

5. DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Identifies clusters in data based on density, effectively handling noise and irregularly shaped clusters.

6. Autoencoders: Neural network models used for feature learning and dimensionality reduction. They can be used for anomaly detection.

7. Isolation Forest: A tree-based ensemble method for identifying anomalies in high-dimensional data.

8. Self-Organizing Maps (SOM): A neural network-based technique for dimensionality reduction and visualization.
