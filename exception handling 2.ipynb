# Q1. Explain why we have to use the Exception class while creating a Custom Exception.

Exceptions are special messages that our program uses to tell us when something unexpected or wrong happens. When we create a custom exception, we are making our own special type of message for situations that are specific to our program.

we use the Exception class as a base when creating custom exceptions beacause it's like having a blueprint that already knows how messages should look and how to handle them. The Exception class is like a set of rules for making and dealing with these special messages.

When we create our custom exception using the Exception class as the base, we are taking advantage of all the helpful things that the Exception class already knows how to do. It's like building on top of something that's already smart about handling messages. This way, we don't have to start from scratch and can focus on making our specific message unique to our program's needs.

By inheriting from the Exception class, we can structure our custom exceptions in a way that makes it easier to manage and handle them appropriately.

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is used as the base class for exceptions that are raised when a key or index used on a mapping or sequence is invalid. It provides a way to handle errors related to accessing elements in a collection, such as dictionaries or lists.

***keyerror***

The KeyError is raised when you try to access a dictionary key that does not exist.

```python
my_dict = {'a': 1, 'b': 2}

try:
    value = my_dict['c']  
except KeyError:
    print("KeyError occurred: Key 'c' not found in dictionary")
```

***IndexError:***
    
The IndexError is raised when you try to access an index that is out of range in a sequence, such as a list or a string.

```python
my_list = [10, 20, 30]

try:
    element = my_list[5]  # Raises IndexError
except IndexError:
    print("IndexError occurred: Index out of range")
```

# Q5. Explain ImportError. What is ModuleNotFoundError?

The ImportError is a built-in exception in Python that is raised when an import statement fails to locate and load a module or a component from a module. This exception occurs when the interpreter encounters difficulties while attempting to import a module or its attributes. Importing modules is a fundamental aspect of Python programming, and the ImportError helps handle various scenarios where imports go wrong.

common situations in ImportError:-

1.Module Not Found
2.Circular Imports
3.Missing Dependencies
4.Incorrect Path

```python
#ModuleNotFoundError
#suppose we have module named my_module.py and it contain following functions


def greet(name):
    print(f"Hello, {name}!")

def farewell(name):
    print(f"Goodbye, {name}!")
```
