# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.graph_objects as go
```

```python
a=sns.load_dataset("titanic")
a
```

```python
fig= go.Figure()
fig.add_trace(go.Scatter(x=a.age, y=a.fare,mode='markers' ))

fig.update_layout(
    title='scatter Plot of Age Vs fare',
    xaxis_title='Age',
    yaxis_title='fare'
)

fig.show()
```

# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
b=sns.load_dataset("tips")
b
```

```python
fig=go.Figure()
fig.add_trace(go.Box(x=b.day, y=b.total_bill,))

fig.update_layout(
    title='Box Plot of Total Bill Amounts by Day',
    xaxis_title='Day of the Week',
    yaxis_title='Total Bill Amount'
)
fig.show()
```

# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
b
```

```python
import plotly.express as px
```

```python

fig = px.histogram(
    b,
    x="sex",
    y="total_bill",
    color="day",
    barmode='group'
)

fig.update_layout(
    title="Histogram of Total Bill by Sex",
    xaxis_title="Sex",
    yaxis_title="Total Bill",
    legend_title="Day"
)

fig.show()
```

# Screenshot


```python

fig = px.histogram(
    b,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker"
)

fig.update_layout(
    title="Histogram of Total Bill by Sex",
    xaxis_title="Sex",
    yaxis_title="Total Bill",
    legend_title="Day"
)

# fig.show()
```

# Q5. What is Distplot? Using Plotly express, plot a distplot.

A distplot is a distribution plot which is used to visualize the distribution of a univariate dataset.

It combines a histogram with a kernel density estimate (KDE) plot to provide a smooth representation of the underlying probability distribution of the data.

To create a distplot using Plotly Express, we can use the px.histogram function and specify the marginal parameter as 'rug' or 'box' to add the KDE curve or a box plot to the histogram.

```python
b
```

```python
fig=px.histogram(b,
                 x="total_bill",
                 marginal="rug",
                 color="sex",
                 title="Distribution of Total Bill"
                 )

fig.show()
```
