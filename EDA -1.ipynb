# Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.

1. Fixed Acidity: This feature represents the fixed (non-volatile) acidity in the wine, which is primarily due to the presence of acids like tartaric acid. Fixed acidity can influence the overall taste and balance of the wine. Wines with the right level of acidity tend to be more palatable.

2. Volatile Acidity: Volatile acidity measures the presence of volatile acids, primarily acetic acid, in the wine. High levels of volatile acidity can lead to off-flavors and spoil the wine's quality. Therefore, lower volatile acidity is generally associated with better wine quality.

3. Citric Acid: Citric acid is a weak organic acid that can provide a refreshing and citrusy flavor to the wine. It contributes to the wine's overall taste and can enhance its quality when present in moderate levels.

4. Residual Sugar: This feature represents the amount of residual sugar left in the wine after fermentation. It influences the wine's sweetness. Some wines are known for their sweetness, while others are dry. The right balance of residual sugar is essential for the desired wine style.

5. Chlorides: Chlorides in wine can come from various sources, including the soil where grapes are grown and winemaking processes. Excessive chloride levels can make the wine taste salty and negatively impact its quality.

6. Free Sulfur Dioxide: Sulfur dioxide is used in winemaking as a preservative and to prevent microbial growth. Free sulfur dioxide helps maintain wine freshness and prevents spoilage. It plays a crucial role in the wine's aging potential and overall quality.

7. Total Sulfur Dioxide: This feature represents both free and bound sulfur dioxide in the wine. Total sulfur dioxide levels can affect the wine's aroma, color, and taste. It's important to have an appropriate balance to avoid unwanted sulfur-related off-flavors.

8. Density: The density of wine can provide insights into its alcohol content and sugar concentration. It is a physical property that influences the wine's mouthfeel and body.

9. pH: The pH level indicates the acidity or basicity of the wine. A lower pH (more acidic) can contribute to wine's freshness, while a higher pH (less acidic) can make the wine taste flat. Balancing pH is crucial for overall wine quality.

10. Sulphates: Sulphates, primarily in the form of potassium sulphate, can act as a nutrient for yeast during fermentation. They can affect the wine's aroma and flavor, and the right concentration is important for quality.

11. Alcohol: The alcohol content in wine is a significant factor in its taste and body. It can contribute to the wine's warmth and mouthfeel. Different wine styles have varying alcohol levels.

12. Quality : This is the target variable, representing the quality of wine on a scale from 0 to 10. It is often used as the dependent variable in predictive modeling, and it indicates the overall quality of the wine as assessed by human tasters.

In predicting the quality of wine, each of these features plays a role in determining the wine's sensory characteristics, balance, and overall appeal. The goal is to find patterns and relationships between these features and wine quality, which can help winemakers and enthusiasts make informed decisions about production and selection. Machine learning models can be trained on this dataset to predict wine quality based on these key features.

### Wine quality dataset does not have missing values.

1. Deletion of Rows (Listwise Deletion):

Advantages: Simple, removes missing data.
 
Disadvantages: Loss of data, potential bias.

2. Mean/Median Imputation:

Advantages: Easy, maintains data structure.

Disadvantages: Ignores variability, introduces bias.

3. Mode Imputations

Advantages: Suitable for categorical data.

Disadvantages: May not represent data well, lacks uncertainty handling.

4. Regression Imputation:

Advantages: Captures relationships, accurate.

Disadvantages: Computationally intensive, assumes linearity

5. K-Nearest Neighbors (K-NN) Imputation:

Advantages: Captures local patterns.

Disadvantages: Sensitive to k, computationally expensive.

6. Multiple Imputation:

Advantages: Realistic estimates, handles uncertainty

Disadvantages: Complex, computationally intensive, requires assumptions.

```python
import scipy.stats as stat
import pylab
import seaborn as sns
import matplotlib.pyplot as plt

def qq_plot(df, feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    plt.title(f'Histogram Plot for : {feature}')
    
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.title(f'Q-Q Plot for : {feature}')
    plt.show()
    
```

```python
## Iterate QQ plot on each of variable 
for col in df.columns:
    print(f'Feature : {col}')
    qq_plot(df,col)
    print('\n'+'='*100 + '\n')
```
