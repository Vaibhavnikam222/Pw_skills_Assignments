# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

PMF and PDF are mathematical concepts used in probability and statistics to describe the probability distribution of a discrete and continuous random variable, respectively. 

**1. Probability Mass Function (PMF):**

PMF is used for discrete random variables which takes on specific, distinct values. The PMF assigns a probability to each possible outcome of the discrete random variable.

The sum of all the probabilities in the PMF must equal 1.

Example: when die is rolled it takes the integers from 1 to 6.There is an equal probability of 1/6 to each of the six possible outcomes.

**2. Probability Density Function (PDF):**

PDF is used for continuous random variables which takes any value within a certain range. Instead of assigning probabilities to specific values, the PDF gives the probability density at each point in the continuous variable's range.

The area under the curve of the PDF over its entire range equals 1.

Example: Consider a continuous random variable representing the height of people in a population. The PDF might look like a bell-shaped curve, such as the normal distribution. 

# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

The Cumulative Distribution Function (CDF) is a fundamental concept in probability and statistics that is used to describe the cumulative probability distribution of a random variable. It provides information about the probability that a random variable takes on a value less than or equal to a specific value.

The CDF is defined for both discrete and continuous random variables.

### Example:-
Consider a fair six-sided die roll, a discrete random variable.

The PMF for this die is- P(X = 1) = 1/6 P(X = 2) = 1/6, P(X = 3) = 1/6,P(X = 4) = 1/6, P(X = 5) = 1/6, P(X = 6) = 1/6

Che CDF for this die:

For X = 1: F(1) = P(X ≤ 1) = P(X = 1) = 1/6
For X = 2: F(2) = P(X ≤ 2) = P(X = 1) + P(X = 2) = 1/6 + 1/6 = 1/3
For X = 3: F(3) = P(X ≤ 3) = P(X = 1) + P(X = 2) + P(X = 3) = 1/6 + 1/6 + 1/6 = 1/2
And so on...

### Why CDF is used:

1. It provides a comprehensive view of the distribution of a random variable, summarizing how probabilities are spread across different values.
2. It allows us to easily calculate probabilities for a range of values, not just a single point.
3. It is often used to find percentiles, which are values that divide the distribution into equal probability segments (e.g., the median is the 50th percentile).
4. It is used in hypothesis testing, confidence interval estimation, and other statistical procedures.

# Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

**Bernoulli Distribution:**
The Bernoulli Distribution is a discrete probability distribution that models a random experiment with only two possible outcomes, typically labeled as "success" and "failure." 
 The distribution is characterized by a single parameter, often denoted as "p," which represents the probability of success.
 
 P (X = x) = P^x * (1-P)^ (1-x)
 
 ### Example:-
Consider flipping a fair coin, where getting heads is considered a "success" (X = 1) and getting tails is a "failure" (X = 0). In this case:

Probability of success (heads), p = 0.5

Probability of failure (tails), 1 - p = 0.5

### Difference between Bernoulli and Binomial Distributions:

**1. Number of Trials:**

**Bernoulli Distribution:** Models a single trial or experiment with only two possible outcomes.

**Binomial Distribution:** Models the number of successes in a fixed number (n) of independent Bernoulli trials.

**2. Random Variable:**

**Bernoulli Distribution:** Has a single random variable (X) representing the outcome of a single trial (0 for failure, 1 for success).

**Binomial Distribution:** Has a random variable (X) representing the number of successes in n trials, where X can take values from 0 to n.

**3. Parameter:**

**Bernoulli Distribution:** Has one parameter, p, representing the probability of success in a single trial.

**Binomial Distribution:** Has two parameters, n (the number of trials) and p (the probability of success in each trial).

**4. Probability Mass Function (PMF):**

**Bernoulli Distribution:** PMF is given by P(X = x) = p^x * (1 - p)^(1 - x) for x = 0 or 1.

**Binomial Distribution:** PMF is more complex and is calculated using the binomial coefficient, which accounts for the number of ways to have x successes in n trials.

**5. Use Cases:**

**Bernoulli Distribution:** Typically used to model single events with binary outcomes (e.g., coin flips, pass/fail ).

**Binomial Distribution:** Used to model the number of successes in a fixed number of independent Bernoulli trials (e.g., number of heads in multiple coin flips, number of defective items in a batch).

# Q7: Explain uniform Distribution with an example.

The Uniform Distribution, also known as the rectangular distribution, is a probability distribution in which all values within a certain range are equally likely to occur. In other words, every possible outcome has the same probability of occurring. The probability density function (PDF) of a uniform distribution is constant within the range of possible values and zero outside that range.

### Example:-
Let's consider the roll of a fair six-sided die. The possible outcomes are integers from 1 to 6. In this case, the uniform distribution applies because each outcome (1, 2, 3, 4, 5, and 6) has an equal probability of 1/6.

# Q8: What is the z score? State the importance of the z score.

The Z-score, also known as the standard score, is a statistical measure that quantifies how far a data point is from the mean of a dataset in terms of standard deviations. It is a standardized value that allows us to compare data points from different datasets or distributions.

  Z= x-$\mu$ \ $\sigma$

### Importance:- 

1. Risk Assessment: In finance, Z-scores can be used to assess the financial health and creditworthiness of a company. It help to evaluate the distance of a company's financial ratios from the mean.

2. Hypothesis Testing:It is used in hypothesis testing, particularly in the context of comparing sample means to population means. When comparing sample means, the Z-test can help determine whether the difference is statistically significant.

3. Standardization: Z-scores standardize data, transforming it into a common scale regardless of the original units of measurement. 

4. Identifying Outliers: Z-scores help identify outliers in a dataset. Data points with Z-scores that are significantly higher or lower than the mean may be considered outliers, suggesting they are unusual or possibly erroneous.

5. Normal Distribution Comparison: In a standard normal distribution (a normal distribution with a mean of 0 and a standard deviation of 1), Z-scores directly correspond to percentiles. For example, a Z-score of 1 corresponds to the 84th percentile, meaning the data point is higher than approximately 84% of the data in the distribution.

6. Students  Assessment: Z-scores are often used in educational assessments to standardize scores and rank students based on their performance relative to the mean and standard deviation of a reference group.

7. Quality Control: In manufacturing and quality control, Z-scores are used to assess whether product measurements fall within acceptable tolerance limits. Deviations beyond a certain Z-score threshold may indicate quality issues.

8. Data Transformation: Z-scores can be applied as a data preprocessing step in various statistical and machine learning techniques, such as clustering and principal component analysis (PCA), to normalize or standardize features.

# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The central limit theorem (CLT) states that the distribution of sample means approximates a normal distribution as the sample size gets larger, regardless of the population's distribution.

It states that if we take repeated random samples from any population and calculate the mean of each sample, the distribution of those sample means will approach a normal distribution as the sample size increases, regardless of the shape of the original population distribution.

**Significance of the Central Limit Theorem:**

1. Foundation of Statistical Inference:A fundamental principle of statistical inference is the Central Limit Theorem. It serves as the foundation for other statistical techniques, such as confidence interval estimation, hypothesis testing, and the use of parametric statistical models.

2. Real-World Applicability: Because it enables statisticians and data analysts to make assumptions about the distribution of sample means even when the population distribution is ambiguous or non-normal, it is extremely applicable in practice. This is especially helpful in situations where it is impractical to collect big samples.

3. Normal Approximation: The CLT provides a basis for approximating the distribution of sample means (or other statistics) by a normal distribution, which simplifies many statistical calculations.

4. Sample Size Determination: It helps determine the appropriate sample size required for making statistical inferences. 

5. Quality Control: In industries like manufacturing, where quality control involves taking samples from production lines, the CLT helps assess the quality of the products by making assumptions about the distribution of sample means.

6. Population Inference: It allows for making inferences about population parameters (such as the population mean) based on the distribution of sample means, even when the population distribution is not known.

# Q10: State the assumptions of the Central Limit Theorem.

###  Assumptions of the Central Limit Theorem :

1. Sample Size: The sample size should be sufficiently large. While there is no fixed minimum sample size requirement, the CLT typically assumes that the sample size (n) is reasonably large, often considered to be greater than 30. Larger sample sizes generally lead to better approximations to a normal distribution.

2. Finite Variance: While the CLT is quite robust, it assumes that the population from which the samples are drawn has a finite variance. This means that the spread or variability within the population is not infinite. In practical terms, this assumption is often satisfied.

3. Independence: The samples must be drawn independently from the population. Each observation within one sample should not be related to or affect the selection or measurement of any other observation in that sample or in other samples. 

4. Random Sampling: The samples should be selected randomly from the population. This means that every individual or data point in the population has an equal chance of being included in the sample. 

5. Population Distribution: The CLT does not impose any specific requirements on the shape or nature of the population distribution. It is applicable to both finite and infinite populations and holds for a wide range of population distributions, including those that are not normally distributed. However, for extremely skewed or heavy-tailed populations, larger sample sizes may be necessary for the CLT to apply effectively.
